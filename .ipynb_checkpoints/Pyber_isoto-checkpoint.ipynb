```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

#make sure they came out right!
#print(city_data.head())
#print(ride_data.head())

# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, on = "city", how = "inner")
merged_data.head()

#group by city and fare
clean_merge = merged_data.drop_duplicates(subset='city', keep='first')

# Display the data table for preview
clean_merge.head()
```

```python
##Bubble Plot of Ride Sharing Data ## look at the mpg activity

grouped_data = merged_data.groupby(['city'])

driver_count = grouped_data['driver_count'].sum()
driver_count

#get counts for cities

#make type into the index!
#clean_merge = clean_merge.set_index('type')
#clean_merge.head()

#look at the graph, get the total number of rides per city and average fare
```
